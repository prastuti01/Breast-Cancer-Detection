## Importing Required Libraries

```python
import pandas as pd              # Used to load and manage datasets (tables)
import numpy as np               # Used for numerical operations
import matplotlib.pyplot as plt  # Used to create graphs and plots
import seaborn as sns            # Used to create better-looking statistical plots

# Used to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Used to scale features so they are on the same range
from sklearn.preprocessing import StandardScaler

# Machine learning algorithms used in this project
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Used to evaluate model performance
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Seting a  theme for plots
sns.set_theme(style="whitegrid", palette="Set2")
```

## Importing Dataset

```python

# Reads the CSV file and stores it in a DataFrame(df)
df = pd.read_csv("breast_cancer_dataset.csv")

# Displaying the first 5 rows of the dataset
df.head()
```

## Data Cleaning & Preprocessing

```python
# Removing the 'id' column because it is not useful for prediction
# Removing 'Unnamed: 32' because it is completely empty
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

```python
# Converting diagnosis values into numbers
# M (Malignant) is converted to 1
# B (Benign) is converted to 0
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# Checking if there are any missing values in each column
df.isnull().sum()
```

```python
print(df.columns)     # Displaying column names 
```

## FEATURE SELECTION / ENCODING

```python

# Separating input features (X) from output label (y)
X = df.drop('diagnosis', axis=1)   # All features except diagnosis
y = df['diagnosis']                # Diagnosis column only

# Creating a StandardScaler object
scaler = StandardScaler()

# Scaling the feature values so they have similar ranges
X_scaled = scaler.fit_transform(X)

X.head()
y.head()
```

## TRAIN-TEST SPLIT

```python
# Spliting data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,       # Scaled input features
    y,              # Target labels
    test_size=0.2,  # 20% data for testing
    random_state=42 # Ensures same split every time
)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## TRAINING MACHINE LEARNING ALGORITHMS

```python
# Creating a dictionary to store different machine learning models
models = {  
    "Logistic Regression": LogisticRegression(max_iter=1000),   # Logistic Regression model
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5), # KNN model
    "Support Vector Machine": SVC(kernel='rbf'),                # SVM model
    "Decision Tree": DecisionTreeClassifier(random_state=42),   # Decision Tree model
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)  # Random Forest model
}

results = []  # List to store results

for name, model in models.items():  # Looping through models

    model.fit(X_train, y_train)     # Training model
    y_pred = model.predict(X_test)  # Predicting test data

    accuracy = accuracy_score(y_test, y_pred)  # Calculating accuracy
    report = classification_report(y_test, y_pred, output_dict=True)  # Getting metrics

    results.append({  # Saving model results
        "Model": name,
        "Accuracy": accuracy,
        "Precision": report['weighted avg']['precision'],
        "Recall": report['weighted avg']['recall'],
        "F1-Score": report['weighted avg']['f1-score'],
        "Support": report['weighted avg']['support']
    })

    print(f"\n--- {name} ---")                  # Printing model name
    print(f"Accuracy: {accuracy:.4f}")          # Printing accuracy
    print("Classification Report:")             # Printing report title
    print(classification_report(y_test, y_pred))# Printing report
    
    # Creating confusion matrix
    cm = confusion_matrix(y_test, y_pred)     

    plt.figure(figsize=(5, 4))                # Creating figure
    sns.heatmap(                              # Ploting heatmap
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=['Benign', 'Malignant'],
        yticklabels=['Benign', 'Malignant']
    )
    plt.xlabel("Predicted Label")              # X-axis label
    plt.ylabel("Actual Label")                 # Y-axis label
    plt.title(f"Confusion Matrix Heatmap - {name}")  # Plots title
    plt.tight_layout()                         # Adjusts layout
    plt.show()                                 # Shows plot
    plt.close()                                # Closees figure
```

## COMPARE AND INTERPRET RESULTS

```python
# Converting results list into a DataFrame
results_df = pd.DataFrame(results)

# Displaying  the comparison table
results_df
```

```python
# Converting the results table into a grouped bar chart

metrics_df = results_df.melt(
    id_vars='Model',                          # Column that identifies each machine learning model
    value_vars=['Accuracy', 'Precision', 'Recall', 'F1-Score'],  # Metrics to be compared
    var_name='Metric',                        # Name for the metric type column
    value_name='Score'                        # Name for the metric value column
)

# Creating a new figure for the bar chart 
plt.figure(figsize=(10, 6))

# Creating a grouped bar chart to compare multiple evaluation metrics for each machine learning model
sns.barplot(
    x='Model',        # Machine learning models on the x-axis
    y='Score',        # Metric scores on the y-axis
    hue='Metric',     # Different colors for Accuracy, Precision, Recall, and F1-score
    data=metrics_df   # DataFrame containing reshaped metric values
)


plt.title("Comparison of Machine Learning Models Using Multiple Metrics")

# Limiting the y-axis range to highlight differences in high-performing models
plt.ylim(0.9, 1.0)

# Rotating model names for better readability
plt.xticks(rotation=45)

# Labeling the axes
plt.xlabel("Machine Learning Model")
plt.ylabel("Performance Score")

# Displaying legend explaining each metric
plt.legend(title="Metric")

# Displaying the final visualization
plt.show()
```

# Dataset Prediction

```python
# Selecting the best-performing model (SVM)
best_model = models["Support Vector Machine"]

# Selecting one sample from test data
sample = X_test[0].reshape(1, -1)

# Predicting diagnosis for the sample
prediction = best_model.predict(sample)

# Printing result in human-readable form
if prediction[0] == 1:
    print("Prediction: Malignant (Cancer Detected)")
else:
    print("Prediction: Benign (No Cancer Detected)")
```
